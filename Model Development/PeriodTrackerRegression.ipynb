```python
df = df[['ClientID', 'CycleNumber',
         'LengthofCycle', 'MeanCycleLength', 'EstimatedDayofOvulation',
         'LengthofLutealPhase',
         'TotalDaysofFertility',
         'LengthofMenses', 'MeanMensesLength',  'Age', 'Height', 'Weight', 'BMI']]
```

```python
df.head()
```

```python
df['MeanCycleLength'] = df.groupby('ClientID')['MeanCycleLength'].transform('first')
df['MeanMensesLength'] = df.groupby('ClientID')['MeanMensesLength'].transform('first')
df['Age'] = df.groupby('ClientID')['Age'].transform('first')
df['Height'] = df.groupby('ClientID')['Height'].transform('first')
df['Weight'] = df.groupby('ClientID')['Weight'].transform('first')
df['BMI'] = df.groupby('ClientID')['BMI'].transform('first')
```
